```python
import pyttsx3
import speech_recognition as sr
import pandas as pd
```

```python
df = pd.read_csv('Ecommerce Customers.csv')
```

```python
class VoiceAssistant:
    def __init__(self, df):
        self.df = df
        self.engine = pyttsx3.init()
        self.engine.setProperty('voice', 'HKEY_LOCAL_MACHINE\SOFTWARE\Microsoft\Speech\Voices\Tokens\TTS_MS_EN-US_ZIRA_11.0')
        
    def _transform(self):
        r = sr.Recognizer()
        with sr.Microphone() as source:
            print("I am listening...")
            r.energy_threshold = 4000
            r.adjust_for_ambient_noise(source,duration=0.1)
            r.pause_threshold = 0.5
            said = r.listen(source)

            try:
                q = r.recognize_google(said, language='en-US')
                return q

            except sr.UnknownValueError:
                print('Sorry I did not understand')
                return 'I am waiting'

            except sr.RequestError:
                print('Sorry the server is down')
                return 'I am waiting'

            except:
                return "I am waiting"
            
    def _speaking(self, message):
        self.engine.say(message)
        self.engine.runAndWait()
        
    def _whatsup(self):
        self._speaking('''Hi, my name is Zira. I am your personal assistant.
                    How may I help you?''')
        
    def querying(self):
        self._whatsup()

        while True:
            q = self._transform().lower()

            if 'entire data' in q:
                display(self.df)
                continue
            
            elif 'describe' in q:
                display(self.df.describe(include='all'))
                continue
                
            elif 'show' in q or 'features' in q:
                print(list(self.df.columns))
                continue

            elif 'shut' in q or 'down' in q:
                self._speaking('Ok. I am shutting down')
                break
```

```python
va = VoiceAssistant(df)
va.querying()
```
